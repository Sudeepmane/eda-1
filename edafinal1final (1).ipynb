```python
import pandas as pd

# Define the function to cap and floor outliers
def cap_floor_outliers(df, column, lower_quantile=0.01, upper_quantile=0.99):
    # Calculate the lower and upper bounds
    lower_bound = df[column].quantile(lower_quantile)
    upper_bound = df[column].quantile(upper_quantile)
    
    # Cap and floor values
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

df = cap_floor_outliers(df, 'LB', lower_quantile=0.01, upper_quantile=0.99)
df= cap_floor_outliers(df,'ASTV',lower_quantile=0.01,upper_quantile=0.99)
df= cap_floor_outliers(df,'MSTV',lower_quantile=0.01,upper_quantile=0.99)
df= cap_floor_outliers(df,'ALTV',lower_quantile=0.01,upper_quantile=0.99)
df= cap_floor_outliers(df,'MLTV',lower_quantile=0.01,upper_quantile=0.99)
df= cap_floor_outliers(df,'Width',lower_quantile=0.01,upper_quantile=0.99)
```

```python
#box plot visualization after treating the outliers using cap and floor method function
sns.boxplot(df)
```
